```python
import requests
```

```python
base_site = "https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250"
```

```python
response = requests.get(base_site)
```

```python
response.ok
content = response.text
```

```python
from bs4 import BeautifulSoup
```

```python
read = BeautifulSoup(content , "html.parser")
```

```python
read
```

```python
movie_name = [i.find("a").text for i in read.find_all("td" , class_ = "titleColumn")]
```

```python
movie_year = [i.text.strip("()") for i in read.find_all("span" , class_ = "secondaryInfo")]
```

```python
movie_rating = [i.text for i in read.find_all("strong")]
```

```python
import pandas as pd
```

```python
movies = pd.DataFrame({"Name": movie_name, "Year": movie_year,"Movie_Rating":movie_rating})
```

```python
movies.to_excel("IMDB TOP MOVIES.xlsx",index = False)
```
